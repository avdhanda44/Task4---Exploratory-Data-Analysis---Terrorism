### <font color='blue'> Data Cleaning </font>

```python
EDA_df.columns.values
```

```python
#Renaming column for self better understanding
EDA_df.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'Day', 'country_txt':'Country','provstate':'State', 'city':'City', 'attacktype1_txt': 'Attacktype', 'targtype1_txt':'Targettype', 'targsubtype1_txt':'Targetsubtype', 'weaptype1_txt':'Weapon', 'gname':'Group', 'region_txt':'Region'}, inplace=True)
```

```python
#Unnecessary columns Removal
#Creating new DataFrame
df = EDA_df[['Year', 'Month', 'Day', 'Country', 'Region','State', 'City', 'Attacktype', 'Targettype', 'Targetsubtype', 'Group', 'Weapon']]
```

```python
df.head(5)
```

```python
# Checking for the null values 
df.isnull().sum()
```

```python
#Filling Null Values
df['State']=df['State'].fillna('Unknown')
```

```python
df['City']=df['City'].fillna('Unknown')
```

```python
df['Targetsubtype'] = df['Targetsubtype'].fillna('others')
```

```python
df.replace( {"Vehicle (not to include vehicle-borne explosives, i.e., car or truck bombs)": "Vehicle"},inplace = True)
```

```python
print("Total {} countries were noted in the data".format(df['Country'].nunique()))
print("Total {} Region were noted in the data".format(df['Region'].nunique()))
print("Total {} Attacktype were noted in the data".format(df['Attacktype'].nunique()))
print("Total {} Targettype were noted in the data".format(df['Targettype'].nunique()))
print("Total {} Group were noted in the data".format(df['Group'].nunique()))
print("Total {} Weapon were noted in the data".format(df['Weapon'].nunique()))
```

### <font color='blue'> Data Visualization </font>

```python
#Destructive Features Of Attacks
print("Year with the most attacks:",df['Year'].value_counts().idxmax())

plt.figure(figsize=(20,5));
sns.countplot(x=df.Year);
plt.xticks(rotation = 10)
plt.title('Attack in Years')
```

```python
print("Month with the most attacks:",df['Month'].value_counts().idxmax())

plt.figure(figsize=(20,5));
sns.countplot(x=df.Month);
plt.title('Attacks by Month')
```

```python
print("Day with the most attacks:",df['Day'].value_counts().idxmax())

plt.figure(figsize=(20,5));
sns.countplot(x=df.Day);
plt.title('Count of Attack on day')
```
